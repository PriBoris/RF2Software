```python
# import scipy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
startPos = 9000
endPos = 22000
pos = np.arange(7000, 24000, 100)
x = (pos - startPos) / (endPos - startPos)
x = np.clip(x, 0.0, 1.0)
```

```python
x1 = np.zeros(x.size)
for i in range(x.size):
    if x[i] < 0.5:
        x1[i] = 1 - 2*x[i]       
    else:
        x1[i] = -1 + 2*x[i]
```

```python
plt.plot(x, x1)
plt.grid(True)
plt.show()
```

```python
y = np.zeros(x1.size)

k = 0.5
p = 4

for i in range(y.size):
    if x1[i] < k:
        y[i] = 0       
    else:
        y[i] = (((x1[i]-k)/(1-k))**p)

    
    
plt.plot(x, y)
plt.grid(True)
plt.show()
```
